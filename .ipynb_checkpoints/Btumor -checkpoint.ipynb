```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import cv2
import os
import random
import matplotlib.pyplot as plt
import pickle
import tensorflow as tf
from tensorflow import keras
from PIL import Image
from keras.models import Sequential 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.layers import Conv2D,MaxPooling2D,Flatten , Dense ,AveragePooling2D ,Dropout , BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam  # Import the Adam optimizer
from tensorflow.keras.applications import MobileNetV2
```

```python
train_dir = r'D:\Studying\ApplAi_CV\Assignments\finalProject\brain_tumor_dataset'
```

```python
data=[]
for files in os.listdir(train_dir):
    label = files
    path = os.path.join(train_dir,label)
    for sub in os.listdir(path):
        data.append({'filename':os.path.join(path,sub) , 'label':label})
        
df = pd.DataFrame(data)        
```

```python
df
```

```python
train_df , val_df = train_test_split(df,test_size=0.2,random_state=42)
```

```python
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=15,               # Adjust rotation range in degrees
    shear_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,                  # Add zoom augmentation
    brightness_range=[0.8, 1.2],     # Adjust brightness
    fill_mode='nearest',
    horizontal_flip=True,
    vertical_flip=True,               # Add vertical flipping
    channel_shift_range=0.2,         # Add channel (color) shifting for color images
)
val_datagen = ImageDataGenerator(rescale = 1/255)

train_generator = train_datagen.flow_from_dataframe(
    train_df,
    x_col = 'filename',
    y_col = 'label',
    target_size=(150,150),
    batch_size=32,
    class_mode='binary'
)
val_generator = val_datagen.flow_from_dataframe(
    val_df,
    x_col = 'filename',
    y_col = 'label',
    target_size=(150,150),
    batch_size=32,
    class_mode='binary'
)
```

```python
# Visualize training data with labels
plt.figure(figsize=(12, 6))
for i in range(10):  # Display the first 10 images
    img, label = train_generator.next()
    plt.subplot(2, 5, i + 1)
    plt.imshow(img[0])  # Display the image
    plt.title("Tumor" if label[0] == 1 else "No Tumor")
    plt.axis('off')
    
plt.suptitle('Training Data with Labels', fontsize=16)
plt.show()
```

```python
# Augmented images
augmented_images, _ = next(train_generator)
plt.figure(figsize=(12, 6))
for i in range(10):  # Display the first 10 augmented images
   plt.subplot(2, 5, i + 1)
   plt.imshow(augmented_images[i])
   plt.axis('off')
plt.suptitle('Augmented Images', fontsize=16)
plt.show()
```

## testing functions

```python
def classify_single_image(image_path, model):
    img_size = (150, 150)
    input_image = cv2.imread(image_path)
    input_image = cv2.resize(input_image, img_size)
    input_image = input_image / 255.0
    input_image = np.expand_dims(input_image, axis=0)
    predictions = model.predict(input_image, verbose = 0)
    predicted_class = predictions[0][0]
    if predicted_class >= 0.5:
        return "Te-me"  # Tumor detected
    else:
        return "Te-no"  # No tumor detected
```

```python
def classify_batch_images(test_directory, model):
    correct = 0
    wrong = 0
    img_size = (150, 150)
    image_files = [f for f in os.listdir(test_directory) if f.endswith('.jpg') or f.endswith('.png')]
    results = []
    for image_file in image_files:
        image_path = os.path.join(test_directory, image_file)
        result = classify_single_image(image_path, model)
        is_tumor = "Te-me" in image_file  # Check if the file name indicates a tumor image
        prediction_is_tumor = result == "Te-me"  # Check if the prediction is "T-me" (tumor detected)
        if (is_tumor and prediction_is_tumor) or (not is_tumor and not prediction_is_tumor):
            color = "green"  # Correct prediction
            title = "Tumor detected" if result == "Te-me" else "Tumor not detected"
            correct +=1 
        else:
            color = "red"  # Incorrect prediction
            title = "Tumor not detected" if is_tumor else "Tumor detected"
            wrong+=1
        results.append((image_path, title, color,correct,wrong))
    return results
```

# Creating our models

## Xception

```python
pretrained_base_model = keras.applications.Xception(
    weights='imagenet',
    include_top=False, # Drop imagenet classifier on the top
    input_shape=(150,150,3)

)
for layer in pretrained_base_model.layers:
  layer.trainable = False
# Define the input shape
inputs = tf.keras.Input(shape=(150, 150, 3))

# Pass the inputs to pretrained model and set training to false
x = pretrained_base_model(inputs, training=False)

# Add a pooling layer
x = tf.keras.layers.GlobalAveragePooling2D()(x)

# Add a dropout layer for regularization purpose: 0.5 dropout rate is a good default
x = tf.keras.layers.Dropout(0.5)(x)

# Last output dense layer with 1 unit. 
# An activation function is not necessary since the predictions are already logit
output = tf.keras.layers.Dense(1)(x)

# Build a model
model = tf.keras.Model(inputs, output)
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(),
    loss=keras.losses.BinaryCrossentropy(from_logits=True),
    metrics=['accuracy'],
)
history = model.fit(
      train_generator,
      validation_data=val_generator,
        epochs = 15
)
```

## ResNet

```python
model = Sequential()
trainedModel = tf.keras.applications.ResNet50V2(
    include_top=False,
    weights="imagenet",
    input_shape=(150, 150, 3),
    pooling=None,
    classes=1000,
    classifier_activation="softmax",
)
for layer in trainedModel.layers:
    layer.trainable = False
model.add(trainedModel)
model.add(Flatten())
model.add(Dropout(0.5)) 
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # Use 'sigmoid' for binary classification
```

```python
opti = Adam(learning_rate=0.0001)
model.compile(optimizer=opti,  # You can choose an optimizer (e.g., 'adam')
              loss='binary_crossentropy',  # Binary cross-entropy loss for binary classification
              metrics=['accuracy'])  # You can add other metrics as needed

# Define callbacks, if desired (e.g., early stopping, model checkpoint)
early_stopping = EarlyStopping(patience=10, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True)

# Fit the model
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=15,  # You can adjust the number of epochs
    validation_data=val_generator,
    validation_steps=len(val_generator),
    callbacks=[early_stopping, model_checkpoint]  # Add callbacks as needed
)
```

## VGG16

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Define the VGG16 base model with pre-trained weights
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(150, 150, 3))
for layer in base_model.layers:
  layer.trainable = False

model = Sequential()

model.add(base_model)

model.add(Flatten())

model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))  # You can adjust the dropout rate as needed

model.add(Dense(1, activation='sigmoid'))

opti = Adam(learning_rate=0.001)
model.compile(optimizer=opti,
              loss='binary_crossentropy',
              metrics=['accuracy'])
early_stopping = EarlyStopping(patience=10, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model_vgg16.h5', save_best_only=True)

```

## MobileNet

```python
input_shape = (150,150,3)

base_model = MobileNetV2(
            input_shape=input_shape,
            include_top=False,
            weights='imagenet')

#2. Locks all the base model's weights
for layer in base_model.layers:
    layer.trainable = False

#3. Adds a few layers to the end of the model
x = Flatten()(base_model.output)
x = Dense(512, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(1, activation='sigmoid')(x)


transfer_model = tf.keras.Model(base_model.input, x)

optimizer = Adam(learning_rate=0.001)  # Adjust the learning rate as needed

transfer_model.compile(loss='binary_crossentropy',
              optimizer=Adam(),
              metrics=['accuracy'])
```

```python
history = transfer_model.fit(
    train_generator,
    epochs=20,
    validation_data=val_generator,
)
```
